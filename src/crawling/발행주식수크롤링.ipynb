```python
import dart_fss as dart
# from config import api_key


def get_corp_list():
    dart.set_api_key(api_key="b9b8f0e09c29872eb0fd4b713c2e3ae408c9d740")
    corp_list = dart.get_corp_list()
    kospi_list = corp_list.find_by_corp_name(corp_name="", market="Y")
    return kospi_list
kospi_list= get_corp_list()
```

```python
kospi_code_table =[]
for i in kospi_list:
    kospi_code_table.append((i.corp_code, i.corp_name))
```

```python
import pandas as pd
# Convert kospi_list to a DataFrame
kospi_list = pd.DataFrame(kospi_code_table, columns=['corp_code', 'company_name'])

# Print the resulting DataFrame
print(kospi_list)
```

```python
import requests
import pandas as pd

api_url = "https://opendart.fss.or.kr/api/stockTotqySttus.json"
crtfc_key = "b9b8f0e09c29872eb0fd4b713c2e3ae408c9d740"  # 인증키를 적절한 값으로 대체해주세요

# 원하는 연도 범위 설정 (2018년부터 2022년까지)
start_year = 2018
end_year = 2022

results = []

for index, row in kospi_list.iterrows():
    corp_code = row["corp_code"]
    company_name = row["company_name"]
    
    for year in range(start_year, end_year + 1):
        url = f"{api_url}?crtfc_key={crtfc_key}&corp_code={corp_code}&bsns_year={year}&reprt_code=11011"
        
        try:
            response = requests.get(url)
            data = response.json()
            
            # 가져온 데이터 활용
            istc_totqy = data["list"][0]["istc_totqy"]
            
            # 결과를 리스트에 추가
            result = [company_name, corp_code, year, istc_totqy]
            results.append(result)
            
        except Exception as e:
            print(f"Error retrieving data for {company_name} ({corp_code}) {year}년:", str(e))
            continue

# 결과 리스트를 DataFrame으로 변환
result_df = pd.DataFrame(results, columns=["Company Name", "Corp Code", "Year", "Issued Shares"])

# 결과 출력
print(result_df)
```
